```python
sys_prompt = prompt1
```

```python
import pandas as pd
train_path = "/content/Astock/data/df_all_year_srl.csv"
df = pd.read_csv(train_path,sep='\t')
```

```python
selected_stock = "贵州茅台"
df_selected = df[df.NAME == selected_stock]
df_selected = df_selected.sort_values("DATE")
df_selected.shape
```

### Connect to ChatGPT

```python
import openai
from google.colab import userdata

openai.api_key = userdata.get('token')

def get_response(x):
    user_prompt=f"假如昨天我看到新闻称:{x.text_a}那么今天，我应当如何投资？请从\'大幅加仓\'、\'小幅加仓\'、\'持有不动\'、\'小幅减仓\'、\'大幅减仓\'中选择一个回答,\
并且阐述你的理由。"
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": sys_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    print(x.DATE)
    print(" - "+ user_prompt)
    print("\n")
    print(" - "+ response.choices[0].message['content'].strip())
    print("\n\n")
    return response
```

```python
df_selected["chatgpt_res"] = df_selected.apply(get_response,axis = 1)
```
